**1. What is hypothesis testing in statistics?**

Ans:
Hypothesis testing is a fundamental statistical method used to determine whether there is enough evidence in a sample of data to support or reject a specific assumption (called a hypothesis) about a population parameter.

**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

Ans:
Null hypothesis:
The null hypothesis (H₀) is a statistical assumption that there is no effect, no difference, or no relationship between variables in a given situation.

Differences b/t Null and Alternative Hypothesis:
Null Hypothesis (H₀):

This is the baseline assumption, often representing "no effect," "no relationship," or "no difference."

It states that any observed changes or effects in the data are purely due to chance or random variation.

Example: "The mean test scores of two groups are equal."

Alternative Hypothesis (H₁):

This hypothesis is what researchers aim to prove; it suggests a presence of an effect, relationship, or difference.

It contradicts the null hypothesis and indicates that the observed data is likely not due to random chance.

Example: "The mean test scores of two groups are different."

**3. What is the significance level in hypothesis testing, and why is it important?**

Ans:
The significance level (α) in hypothesis testing is the threshold used to determine whether the null hypothesis (H₀) should be rejected. It's typically expressed as a probability, such as 0.05 (5%) or 0.01 (1%), representing the risk of rejecting the null hypothesis when it is actually true (a Type I error).

Why is it Important?
The significance level is crucial for the following reasons:

Defines the Decision Boundary:

It sets the cutoff for determining whether the evidence in the data is strong enough to reject H₀. For example, if α = 0.05, you would reject H₀ only if the p-value is ≤ 0.05.

Balances Risks:

It helps balance the trade-off between Type I errors (false positive) and Type II errors (false negative). Lowering α reduces the risk of Type I errors but may increase the risk of Type II errors.

Standard for Scientific Research:

A widely accepted α = 0.05 ensures consistency and comparability across studies, making results credible and easier to validate.

**4.  What does a P-value represent in hypothesis testing?**

Ans:
The p-value in hypothesis testing represents the probability of observing results at least as extreme as the ones obtained, assuming the null hypothesis (H₀) is true. It quantifies how unusual or unlikely the observed data would be if there were no effect or difference in the population.

**5. How do you interpret the P-value in hypothesis testin**g?

Ans:
nterpreting the p-value in hypothesis testing involves understanding its role in evaluating the strength of evidence against the null hypothesis (H₀). Here's how you do it:

Small P-value (≤ α):

Indicates strong evidence against H₀.

Suggests that the observed data is unlikely to have occurred by chance.

Decision: Reject the null hypothesis.

Example: If p-value = 0.02 and α = 0.05, you reject H₀, meaning the results are statistically significant.

Large P-value (> α):

Indicates weak evidence against H₀.

Suggests that the observed data could plausibly have occurred due to random variation.

Decision: Fail to reject the null hypothesis.

Example: If p-value = 0.10 and α = 0.05, you retain H₀, meaning there isn't enough evidence to support the alternative hypothesis.

**6. What are Type 1 and Type 2 errors in hypothesis testing?**

Ans:
Type I and Type II errors are the two possible mistakes that can occur during hypothesis testing. They arise when the decision made based on the test results does not align with reality. Here’s how they work:

Type I Error:
Occurs when you reject the null hypothesis (H₀) even though it is actually true.

Type II Error:
Occurs when you fail to reject the null hypothesis (H₀) even though it is actually false.

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

Ans:
The difference between a one-tailed test and a two-tailed test lies in the directionality of the hypotheses and how the test examines the potential outcomes:

One-Tailed Test:
Definition: Tests for an effect in one specific direction (e.g., either greater than or less than a certain value).

Hypotheses:

Null Hypothesis (H₀): There is no effect or difference (e.g., "The mean is ≤ 50").

Alternative Hypothesis (H₁): The effect or difference occurs in one direction (e.g., "The mean is > 50").

Use Case: Appropriate when you have a clear expectation or theory about the direction of the effect.

Two-Tailed Test:
Definition: Tests for an effect in either direction (e.g., greater than or less than a certain value).

Hypotheses:

Null Hypothesis (H₀): There is no effect or difference (e.g., "The mean = 50").

Alternative Hypothesis (H₁): The effect or difference occurs in either direction (e.g., "The mean ≠ 50").

Use Case: Appropriate when no specific direction of the effect is predicted.


**8. What is the Z-test, and when is it used in hypothesis testing?**

Ans:
The Z-test is a statistical method used to determine whether there is a significant difference between sample data and a population parameter, or between two samples. It is based on the Z-distribution, a special case of the normal distribution.

When is a Z-test Used?

A Z-test is appropriate when:

Sample Size is Large (n ≥ 30): The sample size should be sufficiently large for the Central Limit Theorem to apply, ensuring that the sampling distribution of the mean is approximately normal.

Known Population Standard Deviation (σ): It assumes that the population standard deviation is known, which is a key difference between the Z-test and the t-test.

Hypotheses Involving the Mean or Proportion: It's commonly used for testing hypotheses about population means or proportions.

**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**

Ans:
Calculating the Z-score:
The Z-score formula is:
Z = (X - μ) / σ
Where:
X is the value you're interested in (e.g., a sample mean or a data point)
μ (mu) is the population mean
σ (sigma) is the population standard deviation

What the Z-score Represents:
The Z-score indicates the position of the observed data in the standard normal distribution:

Positive Z-score: Data is above the mean.

Negative Z-score: Data is below the mean.

Z = 0: Data is exactly equal to the mean.

In hypothesis testing, the Z-score helps you evaluate how extreme the observed result is under the null hypothesis. It can be compared to critical Z-values (from the Z-table) to determine statistical significance.

**12. What is the T-test, and how is it used in hypothesis testing?**

The test calculates a t-statistic, which is then used to determine the p-value.

Using it in hypothesis testing:

Formulate the null and alternative hypotheses:

Alternative hypothesis (H1): μ1 ≠ μ2 (difference between means)

Look up the critical t-value or calculate the p-value: Use a t-distribution table or software to find the critical t-value or calculate the p-value.

Interpret the results:
If the p-value is below a certain significance level (e.g., 0.05), reject the null hypothesis, indicating a statistically significant difference between the means.

If the p-value is above the significance level, fail to reject the null hypothesis, indicating no statistically significant difference between the means.

**14. What is a confidence interval, and how is it used to interpret statistical results?**

Ans:
A confidence interval (CI) is a range of values used to estimate a population parameter (such as the mean or proportion) based on sample data. It provides a way to express the degree of uncertainty associated with the estimate and includes the likelihood that the true value falls within this range.

How Is It Used?

Estimation:

It gives researchers an interval to estimate population parameters rather than a single-point value, capturing the inherent uncertainty in sample data.

Testing Hypotheses:

Confidence intervals can complement hypothesis testing. For instance, if the CI excludes the null hypothesis value, it provides evidence to reject it.

Decision-Making:

Confidence intervals provide information about both the magnitude and the precision of an effect, making them useful for practical applications like medical research, policy-making, or quality control.

**15. What is the margin of error, and how does it affect the confidence interval?**

Ans:
The margin of error quantifies the range of uncertainty in an estimate. It reflects the maximum amount by which the sample statistic (e.g., mean or proportion) is expected to differ from the true population parameter due to sampling variability. In the context of a confidence interval, the margin of error defines the width of the interval.

How does the margin of error affect the confidence interval?
A confidence interval is a range of values within which a population parameter is likely to lie. The margin of error is used to construct the confidence interval. The interval is calculated as:
Confidence Interval = (Sample Estimate) ± (Margin of Error)
The margin of error determines the width of the confidence interval. A smaller margin of error results in a narrower confidence interval, indicating more precision in the estimate. A larger margin of error results in a wider confidence interval, indicating less precision.

**16. How is Bayes' Theorem used in statistics, and what is its significance?**

Ans:
What is Bayes' Theorem?
Bayes' Theorem is a mathematical formula that describes how to update the probability of a hypothesis based on new evidence or data.

Formula:
P(H|E) = P(E|H) * P(H) / P(E)
Where:
P(H|E) is the posterior probability of the hypothesis (H) given the evidence (E)
P(E|H) is the likelihood of the evidence given the hypothesis
P(H) is the prior probability of the hypothesis
P(E) is the probability of the evidence

How is Bayes' Theorem used in statistics?

Bayesian inference: Bayes' Theorem is used to update the probability of a hypothesis based on new data. This approach is particularly useful when dealing with complex models or limited data.

Hypothesis testing: Bayes' Theorem can be used to calculate the probability of a hypothesis given the data, allowing for more nuanced testing of hypotheses.

Parameter estimation: Bayes' Theorem can be used to estimate model parameters by updating the probability distribution of the parameters based on the data.

Prediction: Bayes' Theorem can be used to make predictions by updating the probability distribution of the outcome variable based on the data.

**18. What is the Chi-square goodness of fit test, and how is it applied?**

Ans:
The Chi-square goodness of fit test is used to determine whether the observed frequencies in categories match the expected frequencies based on a theoretical distribution. This test helps assess whether data fits a particular model or distribution.

Steps to Apply the Chi-square Goodness of Fit Test:
State the Hypotheses:

Null hypothesis (H₀): The observed distribution matches the expected distribution.

Alternative hypothesis (H₁): The observed distribution does not match the expected distribution.

Calculate the Test Statistic:

Formula: Chi-square = sum of ((observed - expected)^2 / expected)

Where:

observed: Actual frequency of data in each category.

expected: Expected frequency based on the theoretical model.

Find the Degrees of Freedom:

Degrees of freedom = number of categories - 1.

Determine the Critical Value:

Use a Chi-square table to find the critical value based on the significance level (e.g., 0.05) and degrees of freedom.

Compare the Test Statistic to the Critical Value:

If the test statistic exceeds the critical value, reject the null hypothesis.

Otherwise, fail to reject the null hypothesis.

**20. What is an ANOVA test, and what are its assumptions?**

Ans:
Definition:
An ANOVA test (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine if there are significant differences between them.

Assumptions of ANOVA:
ANOVA relies on the following assumptions:

Normality:

The data in each group should be approximately normally distributed.

Homogeneity of Variances:

The variances of the groups should be roughly equal. This is also called the assumption of "homoscedasticity."

Independence:

The observations must be independent of each other. For example, data collected from one participant should not influence data from another.

Equal Sample Sizes (optional):

While ANOVA can handle unequal sample sizes, having equal sizes ensures more accurate results.

**21. What are the different types of ANOVA tests?**

Ans:
Different types of ANOVA:

1. One-Way ANOVA:
Purpose: Compares the means of three or more groups based on a single independent variable (or factor).

2. Two-Way ANOVA:
Purpose: Compares the means across groups based on two independent variables (or factors). It also evaluates if there is an interaction effect between the two variables.

3. Repeated Measures ANOVA:
Purpose: Compares the means of three or more groups where the same participants are measured multiple times (e.g., over time or under different conditions).

4. MANOVA (Multivariate Analysis of Variance):
Purpose: Extends ANOVA to compare multiple dependent variables across groups. It analyzes if the groups differ on a combination of dependent variables.

5. ANCOVA (Analysis of Covariance):
Purpose: Similar to ANOVA, but includes one or more covariates (continuous variables) to control for their influence on the dependent variable.

6. Mixed-Design ANOVA:
Purpose: Combines repeated measures and between-groups designs. It evaluates the effects of both independent groups and repeated measures within the same study.

**3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

```python
def bernoulli_pdf(x, p):

    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    # PDF formula: P(X=x) = p^x * (1-p)^(1-x)
    return (p ** x) * ((1 - p) ** (1 - x))

# Example usage
x = 1  # Observed value
p = 0.7  # Probability of success
print(f"The probability of observing {x} is {bernoulli_pdf(x, p)}")
```

**16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    """
    Calculate and plot the standard normal distribution (mean = 0, std = 1).
    """
    # Generate x values
    x = np.linspace(-4, 4, 1000)  # Range for standard normal distribution
    # Calculate the PDF values
    pdf = norm.pdf(x, 0, 1)

    # Plot the standard normal distribution
    plt.plot(x, pdf, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x, pdf, alpha=0.2, color='blue')  # Add shaded region under the curve
    plt.xlabel('Values (x)')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution')
    plt.axvline(0, color='red', linestyle='--', label='Mean (μ = 0)')
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
plot_standard_normal_distribution()
```

**19. Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    # Calculate sample mean
    sample_mean = np.mean(sample)
    # Calculate sample size
    n = len(sample)
    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Decision based on the significance level
    if p_value < alpha:
        result = "Reject the null hypothesis"
    else:
        result = "Fail to reject the null hypothesis"

    return z_score, p_value, result


# Example usage
sample = [51, 52, 49, 50, 53, 54, 52, 50]  # Sample dataset
population_mean = 50  # Assumed population mean
population_std = 2    # Assumed population standard deviation
alpha = 0.05  # Significance level

z_score, p_value, result = z_test(sample, population_mean, population_std, alpha)

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Result: {result}")
```

**4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate the standard error
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Determine critical Z-values for two-tailed test
    critical_z = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test

    # Calculate the p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Display results
    print("Z-score:", round(z_score, 4))
    print("P-value:", round(p_value, 4))
    print("Critical Z-value:", round(critical_z, 4))

    # Interpretation
    if abs(z_score) > critical_z:
        print("The result is statistically significant. Reject the null hypothesis.")
    else:
        print("The result is not statistically significant. Fail to reject the null hypothesis.")

    # Visualize decision regions
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.plot(x, y, label="Standard Normal Distribution")
    plt.fill_between(x, y, where=(x <= -critical_z) | (x >= critical_z), color='red', alpha=0.5, label="Rejection Region")
    plt.axvline(z_score, color='blue', linestyle='--', label=f"Z-score ({z_score:.2f})")
    plt.axvline(-critical_z, color='green', linestyle='--', label=f"Critical Z (-{critical_z:.2f})")
    plt.axvline(critical_z, color='green', linestyle='--', label=f"Critical Z (+{critical_z:.2f})")
    plt.legend()
    plt.title("Two-Tailed Z-Test Decision Regions")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.show()

# Example values
sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 50

two_tailed_z_test(sample_mean, population_mean, population_std, sample_size)
```

**11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
def bayesian_inference(prior, likelihood, evidence):
    """
    Perform Bayesian Inference using Bayes' Theorem:
    Posterior = (Likelihood * Prior) / Evidence

    Parameters:
    prior (float): Initial belief about the hypothesis (P(H))
    likelihood (float): Probability of the data given the hypothesis (P(D|H))
    evidence (float): Total probability of the data (P(D))

    Returns:
    float: Posterior probability (P(H|D))
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example:
# A disease has a prevalence of 1% (prior = 0.01)
# The test correctly identifies the disease 95% of the time (likelihood = 0.95)
# The test falsely detects the disease in 5% of cases where it's absent
# Evidence = P(D) = P(Disease) * P(D|Disease) + P(No Disease) * P(D|No Disease)

prior = 0.01  # P(Disease)
likelihood = 0.95  # P(Positive Test|Disease)
false_positive_rate = 0.05  # P(Positive Test|No Disease)
evidence = (prior * likelihood) + ((1 - prior) * false_positive_rate)  # P(Positive Test)

# Compute posterior probability
posterior = bayesian_inference(prior, likelihood, evidence)

print(f"Posterior Probability (P(Disease|Positive Test)): {posterior:.4f}")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data. **bold text**

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test.

    Parameters:
    observed_data (2D array): Contingency table containing observed frequencies.

    Returns:
    2D array: Expected frequencies for each cell in the contingency table.
    """
    # Convert to numpy array for processing
    observed = np.array(observed_data)

    # Calculate row and column totals
    row_totals = np.sum(observed, axis=1)
    col_totals = np.sum(observed, axis=0)
    grand_total = np.sum(observed)

    # Calculate expected frequencies
    expected = np.zeros_like(observed, dtype=float)
    for i in range(observed.shape[0]):
        for j in range(observed.shape[1]):
            expected[i, j] = (row_totals[i] * col_totals[j]) / grand_total

    return expected

# Example usage
observed_data = [[20, 30, 50],
                 [40, 10, 50]]

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Data:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed_data, expected_data, alpha=0.05):
    """
    Perform a Chi-square goodness-of-fit test to compare observed and expected distributions.

    Parameters:
    observed_data (array-like): Observed frequency counts.
    expected_data (array-like): Expected frequency counts.
    alpha (float): Significance level for the test (default = 0.05).

    Returns:
    None
    """
    # Perform Chi-square test
    chi2_stat, p_value = stats.chisquare(f_obs=observed_data, f_exp=expected_data)

    # Display results
    print("Chi-square Statistic:", round(chi2_stat, 4))
    print("P-value:", round(p_value, 4))

    # Interpretation
    if p_value < alpha:
        print("The result is statistically significant. Reject the null hypothesis.")
        print("This indicates that the observed distribution differs significantly from the expected distribution.")
    else:
        print("The result is not statistically significant. Fail to reject the null hypothesis.")
        print("This indicates that the observed distribution aligns with the expected distribution.")

observed_data = np.array([50, 30, 20])

# Expected data: Frequencies expected under a hypothesized distribution
expected_data = np.array([40, 40, 20])

goodness_of_fit_test(observed_data, expected_data)
```

**16. Implement an F-test using Python to compare the variances of two random samples.**

```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2, alpha=0.05):

    # Calculate variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate the F-statistic (larger variance divided by smaller variance)
    if var1 > var2:
        f_stat = var1 / var2
        df1 = len(sample1) - 1  # Degrees of freedom for sample1
        df2 = len(sample2) - 1  # Degrees of freedom for sample2
    else:
        f_stat = var2 / var1
        df1 = len(sample2) - 1
        df2 = len(sample1) - 1

    # Calculate the critical value and p-value
    p_value = stats.f.cdf(f_stat, df1, df2) if f_stat < 1 else 1 - stats.f.cdf(f_stat, df1, df2)

    # Display results
    print("F-statistic:", round(f_stat, 4))
    print("P-value:", round(p_value, 4))

    # Interpretation
    if p_value < alpha:
        print("The result is statistically significant. Reject the null hypothesis.")
        print("This indicates that the variances of the two samples are significantly different.")
    else:
        print("The result is not statistically significant. Fail to reject the null hypothesis.")
        print("This indicates that the variances of the two samples are not significantly different.")

# Example Usage
np.random.seed(42)  # Set seed for reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Sample 1
sample2 = np.random.normal(loc=60, scale=15, size=30)  # Sample 2

f_test(sample1, sample2)
```

**19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def check_anova_assumptions(groups):
    print("Checking ANOVA Assumptions...\n")

    # 1. Normality Assumption
    print("1. Checking for Normality within each group:")
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)  # Shapiro-Wilk test for normality
        print(f"  Group {i}: Shapiro-Wilk Test Statistic = {round(stat, 4)}, P-value = {round(p_value, 4)}")
        if p_value > 0.05:
            print("   -> The data appears to be normally distributed.")
        else:
            print("   -> The data does not appear to be normally distributed.")
    print()

    # 2. Equal Variance (Homoscedasticity) Assumption
    print("2. Checking for Equal Variance:")
    stat, p_value = stats.levene(*groups)  # Levene's test for homogeneity of variances
    print(f"  Levene's Test Statistic = {round(stat, 4)}, P-value = {round(p_value, 4)}")
    if p_value > 0.05:
        print("   -> The groups have equal variances (homoscedasticity).")
    else:
        print("   -> The groups do not have equal variances (violates homoscedasticity).")
    print()

    # 3. Independence Assumption
    print("3. Checking for Independence:")
    print("   -> Independence is typically ensured by study design (e.g., random sampling).")
    print("      Please ensure that the data was collected independently.")
    print()

# Example Usage
np.random.seed(42)  # Set seed for reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)  # Group 1
group2 = np.random.normal(loc=55, scale=12, size=30)  # Group 2
group3 = np.random.normal(loc=60, scale=15, size=30)  # Group 3

# Check assumptions
check_anova_assumptions([group1, group2, group3])
```

**24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.**

```python
import numpy as np
import scipy.stats as stats

def chi_square_test_for_variance(sample_data, hypothesized_variance, alpha=0.05):
    # Calculate sample size, sample variance, and test statistic
    sample_size = len(sample_data)
    sample_variance = np.var(sample_data, ddof=1)  # ddof=1 for unbiased sample variance
    chi2_stat = (sample_size - 1) * sample_variance / hypothesized_variance

    # Degrees of freedom
    df = sample_size - 1

    # Calculate critical values for two-tailed test
    critical_lower = stats.chi2.ppf(alpha / 2, df)
    critical_upper = stats.chi2.ppf(1 - alpha / 2, df)

    # Calculate p-value
    p_value = 2 * min(stats.chi2.cdf(chi2_stat, df), 1 - stats.chi2.cdf(chi2_stat, df))

    # Display results
    print("Sample Variance:", round(sample_variance, 4))
    print("Chi-square Statistic:", round(chi2_stat, 4))
    print(f"Critical Values: ({round(critical_lower, 4)}, {round(critical_upper, 4)})")
    print("P-value:", round(p_value, 4))

    # Interpretation
    if chi2_stat < critical_lower or chi2_stat > critical_upper:
        print("\nThe result is statistically significant. Reject the null hypothesis.")
        print("This indicates that the population variance significantly differs from the hypothesized variance.")
    else:
        print("\nThe result is not statistically significant. Fail to reject the null hypothesis.")
        print("This indicates that the population variance does not significantly differ from the hypothesized variance.")

# Example Usage
np.random.seed(42)  # Set seed for reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Simulated sample data
hypothesized_variance = 100

chi_square_test_for_variance(sample_data, hypothesized_variance)
```

**25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.**

```python
import numpy as np
import scipy.stats as stats

def z_test_proportions(success1, size1, success2, size2, alpha=0.05):
    # Calculate the sample proportions
    p1 = success1 / size1
    p2 = success2 / size2

    # Calculate the pooled proportion
    pooled_p = (success1 + success2) / (size1 + size2)

    # Calculate the standard error
    standard_error = np.sqrt(pooled_p * (1 - pooled_p) * (1/size1 + 1/size2))

    # Calculate the Z-score
    z_score = (p1 - p2) / standard_error

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Display results
    print("Proportion 1 (p1):", round(p1, 4))
    print("Proportion 2 (p2):", round(p2, 4))
    print("Z-score:", round(z_score, 4))
    print("P-value:", round(p_value, 4))

    # Interpretation
    if p_value < alpha:
        print("\nThe result is statistically significant. Reject the null hypothesis.")
        print("This indicates that the two proportions are significantly different.")
    else:
        print("\nThe result is not statistically significant. Fail to reject the null hypothesis.")
        print("This indicates that the two proportions are not significantly different.")

# Example Usage
success1 = 40  # Number of successes in group 1
size1 = 100    # Total size of group 1
success2 = 30  # Number of successes in group 2
size2 = 100    # Total size of group 2

z_test_proportions(success1, size1, success2, size2)
```
